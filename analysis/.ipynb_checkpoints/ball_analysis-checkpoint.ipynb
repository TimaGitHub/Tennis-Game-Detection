```python
import pandas as pd
from pathlib import Path
directory_path = "C:\\Users\\Acer\\Downloads\\Dataset\\Dataset\\"
directory_path = Path(directory_path)
csvs = list(directory_path.rglob('*.csv'))
```

```python
all_data = pd.DataFrame([])
```

```python
def re(df):
    df = df.copy()
    df['vx'] = df['x'].diff()[1:]
    df['vy'] = df['y'].diff()[1:]
    df['v'] = np.sqrt(df['vx']**2 + df['vy']**2)
    df['ax'] = df['vx'].diff()[1:]
    df['ay'] = df['vy'].diff()[1:]

    df['x_prev'] = df['x'].shift(1)
    df['y_prev'] = df['y'].shift(1)
    df['vx_prev'] = df['vx'].shift(1)
    df['vy_prev'] = df['vy'].shift(1)
    df['ax_prev'] = df['ax'].shift(1)
    df['ay_prev'] = df['ay'].shift(1)
    df['x_prev_2'] = df['x'].shift(2)
    df['y_prev_2'] = df['y'].shift(2)
    df['vx_prev_2'] = df['vx'].shift(2)
    df['vy_prev_2'] = df['vy'].shift(2)
    df['ax_prev_2'] = df['ax'].shift(2)
    df['ay_prev_2'] = df['ay'].shift(2)

    # Положение, скорость и ускорение на следующем шаге времени
    df['x_next'] = df['x'].shift(-1)
    df['y_next'] = df['y'].shift(-1)
    df['vx_next'] = df['vx'].shift(-1)
    df['vy_next'] = df['vy'].shift(-1)
    df['ax_next'] = df['ax'].shift(-1)
    df['ay_next'] = df['ay'].shift(-1)
    df['x_next_2'] = df['x'].shift(-2)
    df['y_next_2'] = df['y'].shift(-2)
    df['vx_next_2'] = df['vx'].shift(-2)
    df['vy_next_2'] = df['vy'].shift(-2)
    df['ax_next_2'] = df['ax'].shift(-2)
    df['ay_next_2'] = df['ay'].shift(-2)
    df = df.dropna()
    
    return df
    
```

```python
for csv in csvs:
    data = pd.read_csv(csv)
    data = data.drop(columns = 'file name')
    data = data.dropna()
    data.status = data.status.astype(int)
    data = data[data['visibility'] == 1]
    data = data.drop(columns = 'visibility')
    data = data.rename(columns = {'x-coordinate': 'x', 'y-coordinate':'y'})
    data = re(data)
    
    all_data = pd.concat([data, all_data])
    
all_data = all_data.reset_index(drop=True)
```

```python
import lightgbm as lgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
```

```python
all_data_0 = all_data[all_data.status == 0].sample(frac=1)[:5000]
all_data_1 = all_data[all_data.status == 1]
all_data_2 = all_data[all_data.status == 2]
all_data_ = pd.concat([all_data_0,all_data_2, all_data_1])
```

```python
X = all_data_.drop(columns = 'status')
y = all_data_['status']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle = True)
```

```python
model = lgb.LGBMClassifier(class_weight='balanced', verbose = 0)
model.fit(X_train, y_train)

# Оценка модели
accuracy = model.score(X_test, y_test)
print(f"Model accuracy: {accuracy}")
```

```python
y_pred = model.predict(X_test)
report = classification_report(y_test, y_pred)
print(report)
```

```python
y_pred = model.predict(X_test)
report = classification_report(y_test, y_pred)
print(report)
```

```python
model = lgb.LGBMClassifier(is_unbalance = True, verbose = 0)
model.fit(X_train, y_train)

# Оценка модели
accuracy = model.score(X_test, y_test)
print(f"Model accuracy: {accuracy}")
```

```python
y_pred = model.predict(X_test)
report = classification_report(y_test, y_pred)
print(report)
```
